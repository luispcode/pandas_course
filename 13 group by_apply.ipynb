```python
def mean_eur2usd(x):
    return np.mean(x)*1.12
```

```python
mean_eur2usd([100,102])
```

```python
df.groupby('sex')[['total_bill', 'prct_tip']].apply(mean_eur2usd) # aplicar funciones a grupos y columnas utilizando apply y pasandole la funcion creada o build in.
```

```python
df.groupby(['sex', 'time'])[['total_bill', 'prct_tip']].apply(mean_eur2usd)
```

```python
df.groupby(['sex', 'time'])[['total_bill', 'prct_tip']].apply(np.std) # le paso una funcion creada o build in al apply.
```

```python
df.groupby(['sex', 'time'])[['total_bill', 'prct_tip']].apply(lambda x: np.mean(x)*1.12) # definir una funcion en la misma linea con lambda functions.
```

```python
df.groupby(['sex', 'time'])[['total_bill', 'prct_tip']].aggregate([np.mean, np.max]) #aplicar mas de una funcion con aggregate, se puede colocar solo agg tambien.
```

```python
dict_agg = {'tip' : [min, max], 'total_bill': [np.mean, mean_eur2usd]} #definir un diccionario de agregaciones para ser aplicadas.
dict_agg
```

```python
df.groupby(['sex', 'time'])[['total_bill', 'tip']].agg(dict_agg)
```

```python
def f_filter(x):
    return mean_eur2usd(x['total_bill'].mean()) > 20 #definir un filtro en una funcion.
```

```python
df_filtered = df.groupby(['sex', 'time']).filter(f_filter) #filter para aplicar filtros.
df_filtered
```

```python
df_filtered.groupby(['sex', 'time']).count()
```

```python
df['ones'] = 1
df
```

```python
df_g = df.groupby(['sex', 'smoker'])[['ones']].sum() # la nueva celda id cuenta como si fuera un id para hacer el unique count.
df_g
```

```python
df_g.groupby(level=0).apply( #transformar de valores numericos a porcentajes.
lambda x:
    x/ x.sum() *100
        
)
```

```python
pd.cut(df['total_bill'], bins=3).value_counts() #crear bins de una columna categorica.
```

```python
pd.cut(df['total_bill'], bins= [3,18,35,60]).value_counts() #definir el ancho de las bins.
```

```python
df['bin_total'] = pd.cut(df['total_bill'], bins=[3,18,35,60])
df.head(3)
```

```python
df.groupby(['time','bin_total'])[['ones']].count().groupby(level=0).apply( #transformar de valores numericos a porcentajes.
lambda x:
    x/ x.sum() *100
        
)
```
