# Algunas opciones de resample
# Alias - Description
B - Business day
D - Calendar day
W - Weekly
M - Month end
Q - Quarter end
A - Year end
BA - Business year end
AS - Year start
H - Hourly frequency
T, min - Minutely frequency
S - Secondly frequency
L, ms - Millisecond frequency
U, us - Microsecond frequency
N, ns - Nanosecond frequency

```python
df_cum = df_time.resample('12h').sum(min_count=1)
df_cum
```

# opciones para rellenar posibles nan:
df_cum.fillna(-1000)

df_cum.ffill()

df_cum.bfill()

```python
df_cum = df_cum.interpolate()
df_cum
```

```python
df_cum['rate'] = 1 - df_cum['Deaths']/df_cum['Confirmed'] #utilizando una fecha como indice
df_cum
```

```python
df_cum = df_cum.reset_index()
df_cum
```

```python
df_cum.groupby(pd.Grouper(key='ObservationDate', freq = 'M'))[['rate']].mean() #Grouper para usar series de tiempo, utilizando una columna tipo tiempo que no es indice.
```

```python
# ['rate'] serie, [['rate']] data frame.
```

```python
sr = df_cum.groupby(pd.Grouper(key='ObservationDate', freq = '1D'))['rate'].mean()
sr.head(7)
```

```python
sr.plot()
```

```python
sr.rolling(window=7).mean().plot() #rolling se usa con series de tiempo con indices tipo tiempo.
```

```python
sr.plot()
sr.rolling(window=7).mean().plot()
sr.rolling(window=14).mean().plot() #rolling se usa con series de tiempo con indices tipo tiempo.
```

```python
sr.rolling(window=14).apply(lambda x: np.std(x)).plot()
```

```python
df.groupby('Country/Region')['Confirmed'].max().sort_values(ascending=False)
```

```python
df_time = df.groupby(['Country/Region',
           pd.Grouper(key='ObservationDate', freq='1D')]).sum()
df_time
```

```python
df_china = df_time.loc['Mainland China',:]
df_china.head()
```

```python
ax = df_china.plot(figsize= (10,7), title = 'co', legend = False, style= ['r-', 'g--', 'b:*'])
ax.legend(['1', 'dos', '3'])
plt.xlabel('Date')
plt.ylabel('People')
plt.show()
```

```python
df_monthly = df_china.resample('M').max()
df_monthly
```

```python
df_monthly.plot(figsize=(10,7), kind='bar', legend=True)
```

```python
df_monthly.plot(figsize=(10,7), kind='bar', legend=True, stacked=True)
```

```python
df_monthly['Traitment'] = df_monthly['Confirmed'] - df_monthly['Deaths'] - df_monthly['Recovered']
df_monthly
```

```python
df_monthly[['Deaths', 'Recovered', 'Traitment']].T.plot(figsize= (10,7), kind = 'pie', subplots=True)
```
