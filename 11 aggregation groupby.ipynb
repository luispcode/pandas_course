```python
import seaborn as sns
df = sns.load_dataset('diamonds')
df.head()
```

```python
df.groupby('cut').mean()
```

```python
df.groupby('cut').median() # todas las columnas por defecto.
```

```python
df.groupby('cut')['carat'].median()#agrupar por una columna.
```

```python
df.groupby('cut')['carat'].count()
```

```python
df.groupby('cut')['carat'].max()
```

```python
for key_group, group in df.groupby('cut'):
    grouped_price = group['price'].mean()
    
    print('cut: {}, Price: {} \n'.format(key_group, grouped_price))
```

```python
df.groupby(['cut', 'color'])['price'].mean().to_frame() # con to.frame() se convierte una serie a dataframe, el mismo resultado se puede lograr al usar doble corchete
```

```python
def mean_kilo(x):
    return np.mean(x)/1000
```

```python
df.groupby(['cut', 'color'])['price'].aggregate(['min', np.mean, max, mean_kilo]).head(10) #colocarle multiples agregaciones al groupby
```

```python
dict_agg = {'carat':[min, max], 'price':[np.mean, mean_kilo]} #definir un diccionario con agregaciones para pasarle a group by.
dict_agg
```

```python
df.groupby(['cut', 'color']).aggregate(dict_agg) # pasasrle un diccionario de agregaciones a groupby con agregaciones a cada columna.
```

```python
def f_filter(x):
    return mean_kilo(x['price']) > 4
```

```python
df.groupby('cut').filter(f_filter)
```

```python
df.groupby('cut').filter(f_filter)['cut'].unique()
```
