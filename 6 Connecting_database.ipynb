```python
import sys
import sqlalchemy as sql
import pandas as pd
import psycopg2
```

```python
database_type = 'postgresql'
user = 'postgres'
password = 'xxxxxxx'
host = '127.0.0.1:5432'
database = 'postgres'
```

```python
conn_string = '{}://{}:{}@{}/{}'.format(
database_type, user, password, host, database)
sql_conn = sql.create_engine(conn_string)
sql_conn
```

```python
query_sql = """ 
    select *
    from sales
    limit 100;
    """
```

```python
df = pd.read_sql(query_sql, sql_conn)
df.head(5)
```
