```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
pd.options.display.float_format = '{:,.3f}'.format
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.groupby(['sex', 'time']) [['total_bill']].mean()
```

```python
df_gp = df.groupby(['sex', 'time']) [['total_bill']].mean().reset_index() #resetear indice para volver a utilizar un indice numerico.
df_gp
```

```python
df_gp.pivot_table(values='total_bill', index='sex', columns='time') # aplico pivot table a un grupo del data frame.
```

```python
df.pivot_table(values='total_bill', index='sex', columns='time') #aplico el pivot table al data frame directamente y obtengo el mismo resultado, por defecto entrega la media.
```

```python
df.pivot_table(values='total_bill', index='sex', columns='time', aggfunc=np.median) #utilizo otro tipo de agregacion.
```

```python
df_pivot = df.pivot_table(values='total_bill', index='sex', columns='time', aggfunc=[np.median, np.std]) #puedo colocar varias agregaciones.
df_pivot
```

```python
df_pivot.unstack() #unstack para deshacer categorias, vuelve a ser filas.
```

```python
df_pivot.unstack().reset_index() #de nuevo se obtiene un data frame.
```
